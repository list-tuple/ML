```python
s1=["Delhi","Chennai","Hyd","Delhi","Mumbai","Chennai","Hyd"]
```

```python
s2=[10,20,10,20,25,30,25]
```

```python
s3=["Delhi","Tamilnadu","Telangana","Delhi","Maharashtra","Tamilnadu","Telangana"]
```

```python
import pandas as pd
```

```python
df=pd.DataFrame({"City":s1,"Temp":s2,"State":s3})
```

```python
d1=df["City"]
```

```python
d2=df["State"]
```

```python
d3=df["Temp"]
```

```python
df_1=pd.get_dummies(d1)
df_2=pd.get_dummies(d2)
```

```python
df_1
```

```python
df_2
```

```python
df_a=pd.concat([df_1,df_2,d3],axis=1)
```
